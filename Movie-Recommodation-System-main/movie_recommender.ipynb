# Movie Recommdation System

There are 3 types of Recommadation System 
1. Content based : Based on the Similarity of the Content 
2. Collaborative Filter based : Based on the interst of the user , like if 2 users have same interst then the system would suggest the same product to one user which it suggested to the other one
3. Hybrid : Combination of both Content based and Collborative filter based!

Preprocessing the Genres

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'

# I need the above data like this :
# ['Action','Adventure','Fantasy','Science Fiction']
```

```python
# def convert(obj):
#     L = []
#     for i in obj:
#         L.append(i['name'])
#     return L
```

Now we can't pass the list to the convert function as it a string list and we need a numeric list, so for that we will convert the string list into a numeric list 

So for that we use ast.literal_eval 

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# we apply the ast.literal eval here in the function
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# We apply the same concept for keywords as we did for genres
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
# For cast we only need 3 actors 
import ast
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        counter+=1
        if(counter > 2):
            break
    return L

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(1)
```

```python
# Now for the Crew, i am only interested in the director so i will only fetch the name of the Director
import ast
def fetch_crew(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if (i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L
movies['crew'] = movies['crew'].apply(fetch_crew)
```

```python
movies.head()
```

```python
# Now see the overview column is a string and rest like cast , crew , genre and keywords are list so i need to convert the string into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

So now all the features for the tag is converted into a list 

```python
movies.head(1)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
# Now we dont need any spaces in the words like if the word is Sam Worthington then then it should be : SamWorthington so the model dont get confused with different names betwen the spaces 
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head(1)
```

```python
# Now just make a new feature tags and concatinate all the genres , keywords , overview , cast and crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
# can convert the list into a string 
new_df['tags'] = new_df['tags'].apply(lambda x :" ".join(x))
```

```python
# also convert all to lowercase 
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df['tags'][0]
```

```python
new_df.head()
```

```python
# Applying Stemming on the tags 
from nltk.stem import PorterStemmer
# Example usage
stemmer = PorterStemmer()
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))  
        
    return " ".join(y) # converting the list back to string  
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

### Performing Bag of Words in the tags feature

```python
# Creating a back of words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, ngram_range=(2,3), stop_words='english')
# ngram_range=(2,3)) --> from bigram to trigram we are going to use all the features 

vector  = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape # (4806,5000) , 4806 movies and har movie ka 5000 dim ka vector hai 
```

```python
cv.get_feature_names_out()
```

#### Calculating the Cosine Similarity for the model to know how much similar movies are!!

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape # har movie ka har movie se distance nikaalna hai 
```

```python
similarity[0]
```

```python
similarity[0].shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])
```

```python
def recommed(movie):
    movie_index = new_df[new_df['title'] == movie].index[0] # so first we will fetch the index of the movie from the dataset
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1]) # reverse=True sorting in reverse for getting the most similar movies phele .... enumerate takki reverse of the sorting ke baad bhi indicies change na ho
    for i in movie_list[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommed('Avatar')
```

```python
recommed('Batman Begins')
```
